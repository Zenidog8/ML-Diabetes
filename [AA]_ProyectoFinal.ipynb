# Marco Teorico

1. Antecedentes: ¿Qué es la diabetes?

La diabetes mellitus es una enfermedad crónica que afecta la forma en que el cuerpo convierte los alimentos en energía. Normalmente, el cuerpo descompone la mayoría de los alimentos en glucosa (azúcar) y los libera en el torrente sanguíneo. Cuando el nivel de azúcar en sangre sube, el páncreas libera insulina, una hormona que permite que el azúcar entre en las células para ser utilizada como energía.

En personas con diabetes, este proceso se ve afectado porque el cuerpo:

  * No produce suficiente insulina, o

  * No puede usar la insulina de manera eficiente (resistencia a la insulina).

Con el tiempo, esto conduce a niveles elevados de glucosa en la sangre, lo cual puede causar daños en órganos y sistemas del cuerpo.
Tipos de diabetes:

  * Diabetes tipo 1: Enfermedad autoinmune. El sistema inmunológico ataca las células del páncreas que producen insulina. Generalmente se diagnostica en la infancia o adolescencia. Requiere insulina de por vida.

  * Diabetes tipo 2: Es el tipo más común. Se desarrolla generalmente en adultos como consecuencia de resistencia a la insulina, y está fuertemente asociada a obesidad, dieta poco saludable y sedentarismo. Puede manejarse con cambios en el estilo de vida y medicación.

  * Diabetes gestacional: Afecta a algunas mujeres durante el embarazo. Aunque suele desaparecer tras el parto, aumenta el riesgo de desarrollar diabetes tipo 2 en el futuro.

Impacto mundial:

Según la Organización Mundial de la Salud (OMS, 2023):

  * Más de 422 millones de personas viven con diabetes en el mundo.

  * Fue responsable de 1.5 millones de muertes en 2019.

  * La mayoría de los casos son de diabetes tipo 2, y muchos de ellos pueden prevenirse.

## Objetivo del análisis

El objetivo de este proyecto es predecir el valor de la hemoglobina glicosilada (glyhb) en pacientes a partir de variables clínicas, demográficas y de comportamiento contenidas en el dataset. En términos prácticos, se trata de un problema de regresión, en el que la variable objetivo es continua y refleja el nivel de glicemia promedio a largo plazo, siendo un indicador clave para el diagnóstico y monitoreo del control glucémico en diabetes.

A diferencia de un problema de clasificación, donde la meta es asignar una etiqueta (por ejemplo, diabético/no diabético), la regresión busca estimar el valor numérico exacto de una variable de interés. Este enfoque permite no solo identificar el riesgo, sino también cuantificar la severidad potencial o la proximidad a los umbrales diagnósticos, aportando mayor granularidad y valor clínico en la toma de decisiones.

La solución de este problema puede ofrecer múltiples beneficios:

* Permite la detección temprana de riesgos: Anticipar aumentos en los niveles de hemoglobina glicosilada puede facilitar intervenciones preventivas antes de que se establezca la enfermedad.

* Personalización del seguimiento médico: Al estimar valores continuos, se pueden ajustar planes de tratamiento y monitoreo de manera precisa para cada paciente.

* Optimización de recursos sanitarios: Identificar pacientes en riesgo creciente contribuye a la priorización de pruebas y atención médica especializada.

# Entrenamiento y Evaluación de los Modelos

Para abordar el análisis del dataset de diabetes, se han seleccionado tres modelos de regresión supervisada. El objetivo es predecir el nivel de hemoglobina glucosilada (glyhb), que indica el control glucémico de una persona, en función de características personales y clínicas. A continuación se describe cada modelo, sus características, ventajas y razones por las cuales es adecuado para este análisis.

**Regresión Lineal**

Es un modelo estadístico clásico que busca encontrar una relación lineal entre una o más variables independientes (predictoras) y una variable dependiente (respuesta). En este caso, intentamos predecir el valor de glyhb a partir de variables como la edad, el peso, la relación cintura-cadera, entre otras.

Características y beneficios:

  * Fácil de interpretar y explicar.

  * Útil como modelo de línea base (baseline).

  * Permite identificar qué variables tienen mayor influencia en el resultado.

  * Computacionalmente eficiente.

Aplicabilidad:

Se espera que este modelo capture relaciones lineales entre las variables predictoras y el nivel de hemoglobina glucosilada. Es útil para comprender el impacto individual de cada variable, aunque puede no ser suficiente si las relaciones son no lineales.

**SVR (Support Vector Regressor) con kernel RBF**

Es una versión del algoritmo de Support Vector Machines adaptada para regresión. El kernel RBF (Radial Basis Function) permite modelar relaciones no lineales entre las variables.

Características y beneficios:

  * Captura patrones complejos no lineales.

  * Robusto ante outliers y ruido moderado.

  * Requiere ajuste cuidadoso de hiperparámetros (como C, epsilon y gamma), pero puede obtener buenos resultados en datasets estructurados.

Aplicabilidad:

Dado que en el dataset podrían existir relaciones no lineales entre las características (por ejemplo, entre obesidad y glicemia), este modelo puede ofrecer un mejor rendimiento que la regresión lineal simple. Sin embargo, no es fácilmente interpretable.

**XGBoost Regressor**

Es un algoritmo de boosting basado en árboles de decisión. Funciona combinando muchos modelos débiles (pequeños árboles) para construir un modelo fuerte. Es uno de los modelos más potentes y populares en competencias de ciencia de datos.

Características y beneficios:

  * Captura interacciones no lineales y relaciones complejas.

  * Incluye manejo automático de valores faltantes.

  * Tiene mecanismos incorporados para evitar el sobreajuste.

  * Ofrece métricas de importancia de variables.

Aplicabilidad:

XGBoost es ideal para datasets tabulares. Puede manejar relaciones complejas y proporcionar predicciones. Aunque menos interpretable que la regresión lineal, sus métricas de importancia permiten comprender qué variables son más relevantes en el modelo.

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

modelos = {
    "SVR (RBF)": SVR(kernel='rbf'),
    "XGBoostReg": XGBRegressor(objective='reg:squarederror', random_state=42),
    "LinearReg": LinearRegression(),
}

resultados = []
```

```python
for name, model in modelos.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    resultados.append({
        "Modelo": name,
        "RMSE": np.sqrt(mean_squared_error(y_test, y_pred)),
        "MAE": mean_absolute_error(y_test, y_pred),
        "R^2 Score": r2_score(y_test, y_pred)
    })

resultados_df = pd.DataFrame(resultados)
resultados_df.head()
```

## Validación

### ¿Qué es la validación cruzada y por qué la aplicamos?

La validación cruzada es una técnica de evaluación utilizada en Machine Learning para medir la capacidad de generalización de un modelo predictivo. Consiste en dividir el conjunto de datos en varios subconjuntos o “folds”. El modelo se entrena varias veces, cada vez usando una combinación diferente de estos subconjuntos como entrenamiento y el restante como prueba, rotando los roles en cada iteración.

¿Por qué la aplicamos?

* Evita la sobreajuste (overfitting): Al evaluar el modelo en distintos subconjuntos, nos aseguramos de que el buen desempeño no sea solo por azar o porque el modelo memorizó los datos de entrenamiento.

* Aprovecha mejor la información: Especialmente útil con conjuntos de datos de tamaño moderado, ya que cada dato es usado tanto para entrenar como para probar el modelo.

* Entrega una métrica más realista: Proporciona una estimación promedio y más estable del desempeño esperable en datos nuevos, mejor que una sola división entrenamiento/prueba.

### Objetivo de la validación cruzada en este contexto:
Obtener una medida confiable de la capacidad predictiva real de los modelos de regresión aplicados (Regresión Lineal, SVR, XGBoost) para predecir el valor de la hemoglobina glicosilada (glyhb). Así, podemos comparar modelos y seleccionar el que ofrezca el mejor equilibrio entre precisión y generalización.

```python
modelos = {
    "SVR (RBF)": SVR(kernel='rbf'),
    "XGBoostReg": XGBRegressor(objective='reg:squarederror', random_state=42),
    "LinearReg": LinearRegression(),
}
```

```python
# Definir función para RMSE manual
def rmse_cv(model, X, y, cv):
    rmse_resultados = []
    for train_idx, test_idx in cv.split(X):
        model.fit(X.iloc[train_idx], y.iloc[train_idx])
        y_pred = model.predict(X.iloc[test_idx])
        rmse = np.sqrt(mean_squared_error(y.iloc[test_idx], y_pred))
        rmse_resultados.append(rmse)
    return np.mean(rmse_resultados)
```

```python
from sklearn.model_selection import KFold, cross_val_score

# KFold cross-validation con 5 particiones
cv = KFold(n_splits=5, shuffle=True, random_state=42)

cv_resultados = []
for name, model in modelos.items():
    resultados = {'Modelo': name}
    # R2
    resultados['R2'] = np.mean(cross_val_score(model, X_scaled, y, scoring='r2', cv=cv))
    # MAE
    resultados['MAE'] = -np.mean(cross_val_score(model, X_scaled, y, scoring='neg_mean_absolute_error', cv=cv))
    # RMSE (manual)
    resultados['RMSE'] = rmse_cv(model, X_scaled, y, cv)
    cv_resultados.append(resultados)
```

```python
cv_resultados_df = pd.DataFrame(cv_resultados)

# Visualización de resultados
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
metrics_titles = {'R2': 'R² (Promedio)', 'MAE': 'MAE (Promedio)', 'RMSE': 'RMSE (Promedio)'}

for i, metric in enumerate(['R2', 'MAE', 'RMSE']):
    axs[i].bar(cv_resultados_df['Modelo'], cv_resultados_df[metric], color='steelblue')
    axs[i].set_title(metrics_titles[metric])
    axs[i].set_xlabel('Modelo')
    axs[i].set_ylabel(metric)
    axs[i].set_ylim(bottom=0 if metric != 'R2' else None)
    if metric == 'R2':
        axs[i].set_ylim(top=1)

plt.suptitle("Comparación de modelos de regresión con validación cruzada (5-Fold)", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
